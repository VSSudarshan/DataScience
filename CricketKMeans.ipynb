```python
#Importing Libraries
import pandas as pd

# For Visualisation
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# To Scale our data
from sklearn.preprocessing import scale

# To perform KMeans clustering 
from sklearn.cluster import KMeans

# To perform Hierarchical clustering
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import cut_tree
```

```python
#reading Dataset
cricket = pd.read_csv("Cricket.csv",  sep = ',',encoding = "ISO-8859-1", header= 0)
cricket.head()
```

```python
cricket = cricket[['Player', 'Ave','SR']]
```

```python
cricket.head()
```

```python
from sklearn.preprocessing import StandardScaler
standard_scaler = StandardScaler()
RFM_norm1 = standard_scaler.fit_transform(cricket[cricket.columns[2:]])
cluster =KMeans(n_clusters=4,random_state=100,max_iter=50)
```

```python
cricket["cluster"] = cluster.fit(RFM_norm1)
```
